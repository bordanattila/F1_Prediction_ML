```python
# Check for missing values
missing_values = conc_df.isnull().sum()
print(f'*** Number of missing values: {missing_values}')
# Check for duplicates
duplicates = conc_df.duplicated().sum()
print(f"*** Number of duplicate rows: {duplicates}")
   
```

```python
# Check rows where values in driverid and teamid columns are missing
missing_rows = conc_df[conc_df['driverid'].isnull() | conc_df['teamid'].isnull()]   
print(f"*** Number of rows with missing driverid or teamid: {len(missing_rows)}")
missing_rows.head(10)
```

```python
# Check rows where values in driverid and teamid columns are missing and sessiontype isn't practice
sessiontype = conc_df['sessiontype'].astype(str).str.strip().str.lower()
missing_rows_not_practice = conc_df[(conc_df['driverid'].isnull() | conc_df['teamid'].isnull()) & (sessiontype != 'practice')] 
print(f"*** Number of rows with missing driverid or teamid: {len(missing_rows_not_practice)}")
missing_rows_not_practice.head(10)
```

## <font color='teal'>Teams allow reserve drivers or future prospects to participate in practice session. Since all missing values are from practice sessions missing driverid is justified. teamid will be filled based on teamname</font>

```python
# Check for missing values in gridposition and status columns
missing_gridposition_and_status = conc_df[(conc_df['gridposition'].isnull() | conc_df['status'].isnull())]
print(f"*** Number of rows with missing gridposition or status: {len(missing_gridposition_and_status)}")
missing_gridposition_and_status.head(10)
```

```python
# Check for missing values in gridposition and status columns where sessiontype isn't practice
sessiontype = conc_df['sessiontype'].astype(str).str.strip().str.lower()
missing_gridposition_and_status = conc_df[(conc_df['gridposition'].isnull() | conc_df['status'].isnull()) & (sessiontype != 'practice')]
print(f"*** Number of rows with missing gridposition or status: {len(missing_gridposition_and_status)}")
missing_gridposition_and_status.head(10)
```

```python
# Check for missing values in gridposition and status columns where sessiontype isn't practice or qualifying
sessiontype = conc_df['sessiontype'].astype(str).str.strip().str.lower()
missing_gridposition_and_status = conc_df[(conc_df['gridposition'].isnull() | conc_df['status'].isnull()) & (sessiontype != 'practice') & (sessiontype != 'qualifying')]
print(f"*** Number of rows with missing gridposition or status: {len(missing_gridposition_and_status)}")
missing_gridposition_and_status.head(10)
```
